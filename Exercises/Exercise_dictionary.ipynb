We have following information on countries and their population (population is in crores),

| Country  | Population |
|----------|-----------:|
| China    | 143        |
| India    | 136        |
| USA      | 32         |
| Pakistan | 21         |


i Using above create a dictionary of countries and its population

ii Write a program that asks user for three type of inputs,

a print: if user enter print then it should print all countries with their population in this format,
                china==>143
                india==>136
                usa==>32
                pakistan==>21

b add: if user input add then it should further ask for a country name to add. If country already exist in our dataset then it should print that it exist and do nothing. If it doesn't then it asks for population and add that new country/population in our dictionary and print it

c remove: when user inputs remove it should ask for a country to remove. If country exist in our dictionary then remove it and print new dictionary using format shown above in (a). Else print that country doesn't exist!

d query: on this again ask user for which country he or she wants to query. When user inputs that country it will print population of that country.

```python
dict_country_population ={"China":143,
                          "India":136,
                          "USA":32,
                          "Pakistan": 21
                         }
dict_country_population
```

```python
def user_inputs_data(input_dat):
    if input_dat =="print":
        for country, people in dict_country_population.items():
            print(f"{country} ==> {people}")
    elif  input_dat =="add":
        a=input("Tell the country")
        if a in dict_country_population:
            print("it is exist and do nothing")
        else:
            b=input("Tell the population")
            dict_country_population.update({a:b})
            for country, people in dict_country_population.items():
                print(f"{country} ==> {people}")
    elif input_dat =="remove":
        a=input("Tell the country to remove")
        if a in dict_country_population:
            del dict_country_population[a]
            for country, people in dict_country_population.items():
                print(f"{country} ==> {people}")
        else:
            print("That country doesn't exist!")
    elif input_dat =="query":
        a=input("Tell the country to query")
        if a in dict_country_population:
            return dict_country_population[a]
        else:
            print("That country doesn't exist!")
```

```python
abc=input("Enter operation (add, remove, query or print):")
user_inputs_data(abc.lower())

```

You are given following list of stocks and their prices in last 3 days,

| Stock  | Prices           |
|--------|------------------|
| info   | [600,630,620]    |
| ril    | [1430,1490,1567] |
| mtl    | [234,180,160]    |


Write a program that asks user for operation. Value of operations could be,
1 print: When user enters print it should print following,
info ==> [600, 630, 620] ==> avg:  616.67
ril ==> [1430, 1490, 1567] ==> avg:  1495.67
mtl ==> [234, 180, 160] ==> avg:  191.33
2 add: When user enters 'add', it asks for stock ticker and price. If stock already exist in your list (like info, ril etc) then it will append the price to the list. Otherwise it will create new entry in your dictionary. For example entering 'tata' and 560 will add tata ==> [560] to the dictionary of stocks.

```python
stock_data ={"info":[600,630,620],
             "ril":[1430,1490,1567],
             "mtl": [234,180,160] 
            }
```

```python
def user_inputs_data(input_dat):
    if input_dat =="print":
        for stock, price in stock_data.items():
            average=sum(price) / len(price)
            print(f"{stock} ==> {price}==> {average:.2f}"  )
    elif  input_dat =="add":
        a=input("stock ticker")
        new_prices = list(map(int, input("Enter new prices (space-separated): ").split()))
        
        if a in stock_data:
            stock_data[a] = new_prices
            for stock, price in stock_data.items():
                average=sum(price) / len(price)
                print(f"{stock} ==> {price}==> {average:.2f}"  )
        else:
            stock_data.update({a:new_prices})
            for stock, price in stock_data.items():
                average=sum(price) / len(price)
                print(f"{stock} ==> {price}==> {average:.2f}"  )
```

```python
abc=input("Enter operation (add, or print):")
user_inputs_data(abc.lower())
```

3. Write circle_calc() function that takes radius of a circle as an input from user and then it calculates and returns area, circumference and diameter. You should get these values in your main program by calling circle_calc function and then print them

```python
def circle_calc(r_value):
    r=r_value
    pi =22/7
    area=pi * (r**2)
    circumference=2 * pi * r
    diameter =2 * r

    print("circle area:",area)
    print("circle circumference:",circumference)
    print("circle diameter:",diameter)

r_value=(input("enter value for radios of a circle"))
r_val=int(r_value)
circle_calc(r_val)
    
```
