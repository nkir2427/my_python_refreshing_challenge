#### Using following list of cities per country,
india = ["mumbai", "banglore", "chennai", "delhi"]
pakistan = ["lahore","karachi","islamabad"]
bangladesh = ["dhaka", "khulna", "rangpur"]



```python
india = ["mumbai", "banglore", "chennai", "delhi"]
pakistan = ["lahore","karachi","islamabad"]
bangladesh = ["dhaka", "khulna", "rangpur"]

print("india_cities:",india)
print("pakistan_cities:",pakistan)
print("bangladesh_cities:",bangladesh)
```

##### 1. Write a program that asks user to enter a city name and it should tell which country the city belongs to


```python
city_name=input("enter city_name")

if city_name in india:
    print("city is in india")
elif city_name in pakistan:
    print("city is in pakistan")
elif city_name in bangladesh:
    print("city is in bangladesh")
else:
    print("city is in anywhere in the world")
```

##### 2. Write a program that asks user to enter two cities and it tells you if they both are in same country or not. For example if I enter mumbai and chennai, it will print "Both cities are in India" but if I enter mumbai and dhaka it should print "They don't belong to same country"


```python
city_name1=input("enter first city")
city_name2=input("enter second city")

if city_name1 in india and city_name2 in india:
    print("Both cities are in India")
elif city_name1 in pakistan and city_name2 in pakistan :
    print("Both cities are in pakistan")
elif city_name1 in bangladesh and city_name2 in bangladesh:
    print("Both cities are in bangladesh")
else:
    print("They don't belong to same country")
```

##### 3. Write a python program that can tell you if your sugar is normal or not. Normal fasting level sugar range is 80 to 100.
###### Ask user to enter his fasting sugar level
###### If it is below 80 to 100 range then print that sugar is low
###### If it is above 100 then print that it is high otherwise print that it is normal

```python
fasting_sugar_leval =input("enter your fasting sugar leval")
fasting_sugar_leval =int(fasting_sugar_leval)
if fasting_sugar_leval <80 :
    print("sugar is low")
elif fasting_sugar_leval >100 :
    print("sugar is high") 
else:
    print("sugar is normal")
```
